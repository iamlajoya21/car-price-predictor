```python
address_dict = {}
not_found_l = []
addresses_df['Հասցե'].values
```

```python
for a in tqdm(addresses_df['Հասցե'].values):
    try:
        URL = f'https://maps.googleapis.com/maps/api/geocode/json?address={a}&key={geo_api}'
        address_dict[a] = response = r.get(URL).json()
        time.sleep(0.1)
    except:
        not_found_l.append(a)
```

```python
address_geo_df = pd.DataFrame(address_dict).transpose()
```

```python
def geo_parser_lng(x):
    try:
        return x[0]['geometry']['location']['lng']
    except:
        return None

def geo_parser_lat(x):
    try:
        return x[0]['geometry']['location']['lat']
    except:
        return None
```

```python
address_geo_df['lat']  = address_geo_df['results'].apply(geo_parser_lat)
address_geo_df['lng']  = address_geo_df['results'].apply(geo_parser_lng)
```

```python
# address_geo_df['district'] = address_geo_df.index
address_geo_df.reset_index(inplace=False)
```
